```python
# Text Processing & Vectorization using TF-IDF
# Computing Similarity using Cosine Similarity to find similar medicines.
```

```python
import random
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
```

```python
# Loading the cleaned dataset
dataset = pd.read_csv('../datasets/medicines_cleaned.csv')
dataset
```

```python
# Find null values
dataset.isnull().sum()
```

```python
# Filling null values into drug content column
dataset['drug_content'] = dataset['drug_content'].fillna('').astype(str)
dataset
```

```python
# TF - IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=6000)
tfidf_matrix = tfidf_vectorizer.fit_transform(dataset['drug_content'])
```

```python
tfidf_matrix
```

```python
# Computing cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python
# Function to get similar medicines
def get_medicines_by_disease(disease_name, top_n=5):
    disease_meds = dataset[dataset['disease_name'] == disease_name].copy()
    if disease_meds.empty:
        return "No medicines found for this disease."
    results = disease_meds[['med_name', 'final_price', 'drug_manufacturer']]
    return results.sample(n=min(top_n, len(results)))
```

```python
def get_unique_recommendations(disease_name, num_meds=5):
    return get_medicines_by_disease(disease_name, num_meds)
```

```python
recommended_medicines = get_unique_recommendations("Angina")
```

```python
# Using Cosine similarity
def get_medicine_recommendation(disease_name, num_meds=5):
    disease_med = dataset[dataset['disease_name'] == disease_name]
    if disease_med.empty:
        return 'No medicines found for this disease.'
    ref_index = random.choice(disease_med.index)
    sim_score = list(enumerate(cosine_sim[ref_index]))
    sim_score = sorted(sim_score, key=lambda x: x[1], reverse=True)
    similar_indices = [i[0] for i in sim_score[1:num_meds+1]]
    return dataset.iloc[similar_indices][['med_name', 'final_price', 'drug_manufacturer']]
```

```python
recommended_medicines = get_medicine_recommendation("Acne", num_meds=5)
recommended_medicines
```

```python
recommended_medicines = get_medicine_recommendation("Acne", num_meds=5)
recommended_medicines
```

```python
# Using KNN with tfidf matrix
knn = NearestNeighbors(n_neighbors=5, metric='cosine')
knn.fit(tfidf_matrix)
```

```python
def get_medicine_recommendation_using_KNN(disease_name, num_meds=5):
    disease_meds = dataset[dataset['disease_name'] == disease_name]
    if disease_meds.empty:
        return "No medicines found for this disease."
    ref_index = random.choice(disease_meds.index)
    distances, indices = knn.kneighbors(tfidf_matrix[ref_index], n_neighbors=num_meds+1)
    similar_meds = indices.flatten()[1:num_meds+1]
    return dataset.iloc[similar_meds][['med_name', 'final_price', 'drug_manufacturer', 'drug_content']]
```

```python
recommended_medicines = get_medicine_recommendation_using_KNN("Acne", num_meds=5)
recommended_medicines
```
